Scatter

```python
import matplotlib.pyplot as plt
x = [5,7,8,7,2,17,2,9,4,11,12,9,6]
y = [99,86,87,88,111,86,103,87,94,78,77,85,86]
plt.scatter(x, y)
plt.show()
```

Random Data Distributions

```python
import numpy
import matplotlib.pyplot as plt
x = numpy.random.normal(5.0, 1.0, 10000)
y = numpy.random.normal(2.0, 4.0, 10000)
plt.scatter(x,y)
plt.show()
```
